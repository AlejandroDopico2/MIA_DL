```python
from keras.layers import Conv2D, Input, Rescaling, MaxPooling2D, Flatten, Dense
from keras.models import Sequential
from keras.optimizers import Adam
from layers import Inception
from classifier import AnimalImageClassifier
from utils import load_data
IMG_SIZE = 200
BATCH_SIZE = 64

# load data
train, dev, test = load_data('animals/', IMG_SIZE, BATCH_SIZE)
```

## Inception-based Model

```python
model = Sequential([
    Input(shape=(IMG_SIZE, IMG_SIZE, 3)),
    Rescaling(1./255),
    Inception((32, 32, 32, 32)),
    Conv2D(64, 5), Conv2D(64, 5), MaxPooling2D(3),
    Conv2D(128, 5), Conv2D(128, 5), MaxPooling2D(3),
    Conv2D(128, 5), Conv2D(128, 5), MaxPooling2D(3),
    Flatten(), Dense(5, activation='softmax')
])
classifier = AnimalImageClassifier('inception-based', model, IMG_SIZE)
optimizer = Adam(learning_rate=1e-4)
classifier.train(train, dev, optimizer, 'results/', batch_size=BATCH_SIZE, epochs=50)
```
